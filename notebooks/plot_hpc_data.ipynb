```python
%matplotlib inline
import os
import sys
from os.path import dirname

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy
import pickle
import gdal
import georasters as gr
import geopandas as gp
def gamma_porp(a,b):
    from scipy import special
    return special.gamma(a)-special.gamma(a)*special.gammainc(a,b)
```

```python
X,Y,CVS = pd.read_pickle('./data/output.p')

from numpy import loadtxt
from scipy.ndimage.filters import gaussian_filter
from matplotlib.pyplot import contour, show

sigma = 1.0 # this depends on how noisy your data is, play with it!
CVS = gaussian_filter(CVS, sigma)
CVS[CVS>0.4] = np.nan
c = plt.contourf(X,Y,(CVS), 20)
plt.xlim([0.08, .15])
# plt.ylim([50,600])
plt.xlabel('Storm frequency [events per day]')
plt.ylabel('Storage capacity [mm]')
plt.title('Dry season storage sensitivity\nto climate and storage capacity')
cb = plt.colorbar()
cb.set_label('CV conditions @ dry season start ')
```
